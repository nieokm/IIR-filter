```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')

data_in=[]
data_out=[]

for num in range(0,3):
    data_in.append( np.loadtxt("input"+str(num)+".txt") )
    data_out.append( np.loadtxt("output"+str(num)+".txt") )
labels=['sinus','twosinus','white noise']
colors=['b', 'r', 'g']
fig = plt.figure(figsize=(15,10))
ax = fig.add_subplot(221)

for num in range(0,2): ##plot for white noise is senseless. it's very messy, since it's not a continious function
    ax.plot(data_in[num], colors[num], label=labels[num])
plt.ylabel(r"$f(x_{N})$", axes=ax)
plt.xlabel(r"$N$", axes=ax)
ax.legend(title='Signal type', bbox_to_anchor=(1, 1), loc='upper left')
plt.title("Input", axes=ax)

ax2 = fig.add_subplot(223)
for num in range(0,3):
    ax2.plot(data_out[num], colors[num], label=labels[num])
plt.ylabel(r"$f(x_{N})$", axes=ax2)
plt.xlabel(r"$N$", axes=ax2)
ax2.legend(title='Signal type', bbox_to_anchor=(1, 1), loc='upper left')
plt.title("Output", axes=ax2)
plt.tight_layout()

fig.savefig('graphs.png', dpi=125, bbox_inches="tight")
plt.show()
```
